# Description

n this notebook, I will extract a table from Wikipedia as a **WEB SCRAPING** project.

This table includes the biggest companies in the world.

# Import Libraries

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# Get Data From the Website

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
page = requests.get(url)
page
```

Find the target table in HTML format

```python
soup = bs(page.text, 'html')
table = soup.find('table')
print(table)
```

# Find Headers

```python
table_header = table.find_all('th')[:10]
table_header = [title.text.strip() for title in table_header]
print(table_header)
```

Some headers should be edited

```python
table_header[-2] = 'Ref'
table_header[-4] = 'Headquarters'
table_header.remove('Rank')
print(table_header)
```

# Find the Data

```python
rows = table.find_all('tr')[2:]
index = []
data = []
for i in range(len(rows)):
    index.append(int(rows[i].find_all('th')[0].text.strip()))
    temp = rows[i].find_all('td')
    hold = [item.text.strip() for item in temp]
    data.append(hold)

print(data)
```

# Create Dataframe


```python
df = pd.DataFrame(data, columns=table_header, index= index)
df
```

Let's remove some additional columns.

```python
df.drop(['State-owned', 'Ref'], inplace=True, axis=1)
```

```python
df.index.name = "Rank"
```

```python
df
```

Finally, we can download the dataset.

```python
file_name = 'Companies.csv'
df.to_csv(file_name)
```
